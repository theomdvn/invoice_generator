```python
import os
import pandas as pd
from tempfile import NamedTemporaryFile
from datetime import datetime
from babel.dates import format_datetime
from InvoiceGenerator.api import Invoice, Item, Client, Provider, Creator

from InvoiceGenerator.pdf import SimpleInvoice
os.environ["INVOICE_LANG"] = "fr"
```

```python
data = pd.read_csv('stockx_historical_sales_report_601fa0f2_c1e5_11ed_9ad6_1220b06a0089_1683559952.csv')

```

```python
client = Client('STOCKX LLC', address='De Run 4312, Veldhoven , 5503LN ',city='TVA intracommunautaire : NL826418247B01',note='1) Intra-EU supply of goods in accordance with article 138 Council Directive 2006/112/EC')
provider = Provider('SAS Deetail',bank_account='Shine',address='88 RUE DANIELE CASANOVA, 91700, SAINTE GENEVIEVE DES BOIS', city = 'TVA : FR95949083174', note='SIRET : 94908317400019 ')
creator = Creator('Theo Midavaine')
```

```python
y = 66
for i in data['Invoice Number']:
    if y < 10:
        y = '0'+str(y)
    invoice = Invoice(client, provider, creator)
    invoice.add_item(Item(count=int(data[data['Invoice Number'] == i]['Quantity'].values[0]),description=str(data[data['Invoice Number'] == i]['Sku Name'].values[0]),price=float(data[data['Invoice Number'] == i]['Total Gross Amount (Total Payout)'].values[0]),tax='0'))
    invoice.generate_breakdown_vat()
    #invoice.generate_breakdown_vat_table()
    invoice.currency = 'EUR'
    invoice.title = str(i)
    invoice.use_tax = True
    invoice.iban = 'FR76 1741 8000 0100 0115 6140 709'
    invoice.number = f'#A2300{y}'
    original_date = (data[data['Invoice Number'] == i]['Invoice Date'].values)[0]
    parsed_date = datetime.strptime(original_date, "%Y-%m-%d %H:%M:%S")

    invoice.date = parsed_date
    invoice.paytype='Virement bancaire'
    pdf = SimpleInvoice(invoice)
    pdf.gen(f"#A2300{y}_{data[data['Invoice Number'] == i]['Total Gross Amount (Total Payout)'].values[0]}.pdf")
    y = int(y)
    y +=1
```
